```python
import pandas as pd
import matplotlib
```

```python
file = 'SP_2010_0301-0331.csv'
```

```python
df = pd.read_csv(file)
print(df.head())
```

```python
print(list(df))
print(len(df))
print(len(list(df)))
```

```python
col = 'Unnamed: 2'
col_name = 'Wind Speed (m/s)'
freq = df[col].value_counts().rename_axis(col_name).reset_index(name='freq')
freq = freq.sort_values(by = col_name)
print(freq)
```

```python
codebook = [['1', 'Plot Title: Sailing'], 
            ['2', 'Unnamed: 1'], 
            ['3', 'Unnamed: 2'],
            ['4', 'Unnamed: 3'],]
for lbl in codebook:
    freq[col_name] = freq[col_name].replace(int(lbl[0]), lbl[1])
print(freq)
```
